```python
import optuna
from pathlib import Path
import sqlite3
import pandas as pd
```

```python
sqlite_file = Path("/Users/thomassimader/Documents/NIRB/db_1.0e-05min_max.sqlite3")
```

```python
# Query to select all rows from the 'studies' table
# Connect to the SQLite database
conn = sqlite3.connect(sqlite_file)
cursor = conn.cursor()
cursor.execute("SELECT * FROM studies")
# Fetch all results
rows = cursor.fetchall()
# Get column names
column_names = [description[0] for description in cursor.description]
# Optionally, display as a DataFrame for better readability
df = pd.DataFrame(rows, columns=column_names)
print(df)
```
